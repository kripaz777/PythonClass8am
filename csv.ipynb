```python
s = ""
b = open('data.csv','w')
b.write("Name,Age,Address\n")
n = int(input("Enter the value of n = "))
for i in range(n):
    name = input("Enter name = ")
    age = int(input("Enter age = "))
    add = input("Enter address = ")
    info = f"{name},{age},{add}\n"
    s = s+info

b.write(s)
b.close()
```

```python
import csv
f = open('bill.csv', 'a')
x = csv.writer(f)
h = ["Name","Price","Quantity","Total"]
x.writerow(h)
data = []
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    price = int(input("Enter price = "))
    quantity = int(input("Enter  = "))
    total = price*quantity
    info = [name,price,quantity,total]
    data.append(info)
    x.writerow(info)
print(data)
f.close()
```

```python
import csv
csv_columns = ['Name', 'Age', 'Address']
dict = [
    {'Name':'Aayush','Age':23,'Address':'Katmandu'},
    {'Name':'Shyam','Age':33,'Address':'Patan'}
]
try:
    with open("data.csv", 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in dict:
            writer.writerow(data)
except IOError:
    print("I/O error")
```
